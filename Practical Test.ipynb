```python
# Python program to demonstrate erosion and
# dilation of images.
import cv2 as cv
import numpy as np
 
# Reading the input image
img = cv.imread('signature2.png', 0)
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((3,3), np.uint8)
 
# The first parameter is the original image,
# kernel is the matrix with which image is
# convolved and third parameter is the number
# of iterations, which will determine how much
# you want to erode/dilate a given image.
img_erosion = cv.erode(img, kernel, iterations=10)
img_dilation = cv.dilate(img, kernel, iterations=1)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)

 
cv.imshow('Input', img)
cv.imshow('Erosion', img_erosion)
cv.imshow('Dilation', img_dilation)
cv.imshow('Opening', opening)
cv.imshow('Closing', closing)
cv.imshow('Morphological Gradient', gradient)
 
cv.waitKey(0)
```

```python
# Python program to demonstrate erosion and
# dilation of images.
import cv2 as cv
import numpy as np
 
# Reading the input image
img = cv.imread('signature2.png', 0)
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((111,111), np.uint8)
 
# The first parameter is the original image,
# kernel is the matrix with which image is
# convolved and third parameter is the number
# of iterations, which will determine how much
# you want to erode/dilate a given image.
img_dilation = cv.dilate(img, kernel, iterations=1)
img_erosion = cv.erode(img, kernel, iterations=4)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)

cv.imshow('Input', img)
cv.imshow('Erosion', img_erosion)
cv.imshow('Dilation', img_dilation)
cv.imshow('Closing', closing)
cv.imshow('Black Hat', blackhat)
 
cv.waitKey(0)
```
